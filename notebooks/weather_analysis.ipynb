```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://weather_user:password123@postgres:5432/weather_db")

```

```python
df_last50 = pd.read_sql_query("""
SELECT id, timestamp, temperature, humidity, pressure, wind_speed
FROM weather_readings
ORDER BY timestamp DESC
LIMIT 50
""", engine)

df_last50
```

```python
df_temp = pd.read_sql_query("""
SELECT timestamp, temperature
FROM weather_readings
WHERE timestamp >= NOW() - INTERVAL '1 hour'
ORDER BY timestamp
""", engine)

df_temp["timestamp"] = pd.to_datetime(df_temp["timestamp"])
df_temp.plot(x="timestamp", y="temperature", figsize=(12,4), title="Temperature last hour")
```

```python
kpi = pd.read_sql_query("""
SELECT
  AVG(temperature) AS avg_temp_1h,
  AVG(pressure)    AS avg_pressure_1h,
  AVG(humidity)    AS avg_humidity_1h,
  MAX(wind_speed)  AS max_wind_1h
FROM weather_readings
WHERE timestamp >= NOW() - INTERVAL '1 hour'
""", engine)

kpi
```
